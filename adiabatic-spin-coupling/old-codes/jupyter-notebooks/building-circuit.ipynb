```python
from qiskit import QuantumCircuit
```

```python
circuit = QuantumCircuit(3,3)
```

```python
circuit.x(1)
```

```python
circuit.h(range(3))
circuit.cx(0,1)
circuit.measure(range(3), range(3));

```

```python
print(circuit)

```

```python
circuit.draw(output="mpl")

```

```python
# Latex drawing
circuit.draw(output="latex")
```
